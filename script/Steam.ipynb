```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

game_list = []
r = requests.get('https://store.steampowered.com/search/?filter=topsellers&os=win')
r
```

```python
soup = BeautifulSoup(r.content, 'lxml')
soup
```

```python
# Find all the game elements in the HTML with the specified class
games = soup.find_all('a', {'class': 'search_result_row ds_collapse_flag'})

# Extract game information from each game element
for game in games:
    game_url = game['href']
    game_img = game.img['src']
    game_name = game.span.text
    game_release = game.find('div', class_='search_released').text
    game_review = game.find('div', class_='search_reviewscore').span['data-tooltip-html'] if game.find('div', class_='search_reviewscore').span else ''
    game_price = game.find('div', class_='search_price').text.strip()
    game_dict = {
        'game_url': game_url,
        'game_img': game_img,
        'game_name': game_name,
        'game_release': game_release,
        'game_review': game_review,
        'game_price': game_price
    }
    game_list.append(game_dict)
```

```python
df = pd.DataFrame(game_list)
df
```
