```python
import requests
from bs4 import BeautifulSoup

login_url = "https://the-internet.herokuapp.com/login"

authenticate_url = 'https://the-internet.herokuapp.com/authenticate'

secure_url = 'https://the-internet.herokuapp.com/secure'

payload = {
    'username': 'tomsmith',
    'password': 'SuperSecretPassword!'
}
```

```python
# Without session
# Send a POST request to the authentication URL with the login credentials
response = requests.post(authenticate_url, data=payload)

# Print the response
print(response)

# Create a BeautifulSoup object to parse the response content
soup = BeautifulSoup(response.content, 'html.parser')

# Print the prettified HTML content
print(soup.prettify())

# Send a GET request to the secure URL
response2 = requests.get(secure_url)

# Print the response
print(response2)

# Create a BeautifulSoup object to parse the response content
soup2 = BeautifulSoup(response2.content, 'html.parser')

# Print the prettified HTML content
print(soup2.prettify())
```

```python
# Create a session object using requests.session()
with requests.session() as s:
    # Send a POST request to the authentication URL with the login credentials
    s.post(authenticate_url, data=payload)

    # Send a GET request to the secure URL within the same session
    response = s.get(secure_url)

    # Print the response
    print(response)

    # Create a BeautifulSoup object to parse the response content
    soup = BeautifulSoup(response.content, 'html.parser')

    # Print the prettified HTML content
    print(soup.prettify())
```
