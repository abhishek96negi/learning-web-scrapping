```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.skysports.com/premier-league-table'

# Send HTTP GET request to the specified URL
response = requests.get(url)

# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find the league table within the HTML using the specified class name
league_table = soup.find('table', class_='standing-table__table callfn')

league_table_point = []

# Iterate over each team in the league table
for team in league_table.find_all('tbody'):
    rows = team.find_all('tr')
    
    # Iterate over each row in the table body
    for row in rows:
        # Extract the team name and points from the respective cells
        p1_team = row.find('a', class_='standing-table__cell--name-link').text
        p1_points = row.find_all('td', class_='standing-table__cell')[9].text
        
        # Create a dictionary to store the team name and points
        p1_data = {
            'Team': p1_team,
            'Points': p1_points
        }
        
        # Append the team data to the league_table_point list
        league_table_point.append(p1_data)  # Store the team data
```

```python
# Convert the list of dictionaries to a pandas DataFrame
league_data_frame = pd.DataFrame(league_table_point)

league_data_frame
```
