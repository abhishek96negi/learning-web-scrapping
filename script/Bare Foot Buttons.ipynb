```python
from requests_html import HTMLSession
import time
import pandas as pd

s = HTMLSession()
url = 'https://barefootbuttons.com/product-category/version-1/'

def get_links(url):
    r = s.get(url)
    
    # Find all the product items using the appropriate HTML selector
    items = r.html.find('div.product-small.box')
    
    links = []
    for item in items:
        # Extract the href attribute from the first 'a' tag found within each item
        links.append(item.find('a', first=True).attrs['href'])
    
    return links

def get_productdata(link):
    r = s.get(link)
    
    # Extract the title, price, tag, and sku from their respective HTML elements
    title = r.html.find('h1', first=True).full_text
    price = r.html.find('span.woocommerce-Price-amount.amount bdi')[1].full_text
    tag = r.html.find('a[rel=tag]', first=True).full_text
    sku = r.html.find('span.sku', first=True).full_text

    # Create a dictionary to store the product data
    product = {
        'title': title.strip(),
        'price': price.strip(),
        'tag': tag.strip(),
        'sku': sku.strip()
    }
    return product

results = []
links = get_links(url)

for link in links:
    # Retrieve the product data for each link and append it to the results list
    results.append(get_productdata(link))
    
    # Add a delay of 1 second before making the next request
    time.sleep(1)

# Create a DataFrame from the results list
df = pd.DataFrame(results)
df
```
