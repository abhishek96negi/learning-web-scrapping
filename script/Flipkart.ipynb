```python
# importing the package
import requests
import pandas as pd

from bs4 import BeautifulSoup # pip install beautifulsoup4 (https://pypi.org/project/beautifulsoup4/)

url = "https://www.flipkart.com/search?q=mac%20pro&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"

response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
print(soup.prettify())
```

```python
product_dict = {
    'PRODUCT_NAME': [],
    'PRODUCT_PRICE': [],
    'PRODUCT_RATING': [],
    'PRODUCT_IMAGE': [],
    'PRODUCT_LINK': [],
}

for rowdiv in soup.find_all('div', class_='_13oc-S'):
    for productdiv in rowdiv.find_all('div', class_='_4ddWXP'):
        product_link = f'https://www.flipkart.com/{productdiv.find("a", class_="_2rpwqI").get("href")}'

        product_name = productdiv.find('a', class_='s1Q9rs').text
        
        product_rating = productdiv.find('div', class_='_3LWZlK').text if productdiv.find('div', class_='_3LWZlK') is not None else ''
        
        product_price = productdiv.find('div', class_='_30jeq3').text
        
        if productdiv.find('img', class_='_396cs4 _3exPp9'):
            product_image = productdiv.find('img', class_='_396cs4 _3exPp9').get('src')
            product_dict['PRODUCT_IMAGE'].append(product_image)
        else:
            product_dict['PRODUCT_IMAGE'].append(None)
            
        product_dict['PRODUCT_NAME'].append(product_name)
        product_dict['PRODUCT_PRICE'].append(product_price)
        product_dict['PRODUCT_RATING'].append(product_rating)
        product_dict['PRODUCT_LINK'].append(product_link)
```

```python
df = pd.DataFrame(product_dict)
df
```
